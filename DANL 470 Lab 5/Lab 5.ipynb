```python
import pandas as pd
import numpy as np
FatSupply=pd.read_csv('FatSupply.csv')
FatSupply
```

```python
FatSupply.corr()
```

```python
FatSupply.drop(['Country', 'Alcoholic Beverages', 'Aquatic Products, Other', 'Cereals - Excluding Beer', 'Fish, Seafood', 'Fruits - Excluding Wine', 'Meat'], axis=1, inplace=True)
FatSupply
```

```python
FatSupply.drop(['Miscellaneous', 'Offals', 'Oilcrops', 'Pulses', 'Spices', 'Starchy Roots', 'Vegetable Oils', 'Vegetables', 'Population', 'Unit (all except Population)'], axis=1, inplace=True)
FatSupply
```

```python
FatSupply.drop(['Sugar Crops', 'Sugar & Sweeteners', 'Treenuts', 'Vegetal Products', 'Obesity', 'Undernourished', 'Confirmed', 'Recovered', 'Active'], axis=1, inplace=True)
FatSupply
```

```python
FatSupply.corr()
```

```python
FatSupply.isnull().sum()
```

```python
FatSupply.Deaths.fillna(FatSupply.Deaths.mean(), inplace=True)
FatSupply.isnull().sum()
```

```python
FatSupply['RevisedDeath'] = np.where(FatSupply['Deaths']>=0.005,'High','Low')
FatSupply.head()
```

```python
FatSupply.drop(['Deaths',],axis=1, inplace=True)
FatSupply.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = FatSupply.drop(['RevisedDeath'], axis=1)
Y = FatSupply['RevisedDeath']
X_train = np.array(X[0:int(0.08*len(X))])
Y_train = np.array(Y[0:int(0.08*len(Y))])
X_test = np.array(X[0:int(0.08*len(X)):])
Y_test = np.array(Y[0:int(0.08*len(Y)):])
len(X_train), len(Y_train), len(X_test), len(Y_test)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
LR = LogisticRegression()
LR_fit = LR.fit(X_train, Y_train)
LR_pred = LR_fit.predict(X_test)
LSVM = LinearSVC()
LSVM_fit = LSVM.fit(X_train, Y_train)
LSVM_pred = LSVM_fit.predict(X_test)
from sklearn.metrics import accuracy_score
print("Logistic Regression is %f percent accurate" % (accuracy_score(LR_pred, Y_test)*100))
print("Linear SVMs is %f percent accurate" % (accuracy_score(LSVM_pred, Y_test)*100))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
seed=0
cv = KFold(n_splits=5,random_state=2, shuffle=True)
def return_score(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
    return score
scores = []
```

```python
model = LogisticRegression()
for train_index, test_index in cv.split(X,Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
    score = return_score(model,X_train, X_test, Y_train, Y_test)
    scores.append(score)
print("Accuracy score in each iteration: {}".format(scores))
print("K-Fold Score: {}".format(np.mean(scores)))
```

```python
model = LinearSVC()
for train_index, test_index in cv.split(X,Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
    score = return_score(model,X_train, X_test, Y_train, Y_test)
    scores.append(score)
print("Accuracy score in each iteration: {}".format(scores))
print("K-Fold Score: {}".format(np.mean(scores)))
```
